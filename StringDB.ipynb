```python
stringDB_Orthology = pd.read_csv('data/4932.protein.orthology.v12.0.txt', sep = '\t', dtype = 'str')
stringDB_Orthology
```

```python
stringDB_COG_links = pd.read_csv('data/COG.links.v12.0.txt',sep = ' ', dtype = 'str')
stringDB_COG_links
```

```python
stringDB_COG_mappings = pd.read_csv('data/COG.mappings.v12.0 .txt', sep = '\t', dtype = 'str')
stringDB_COG_mappings
```

```python
# flter mappings data to only contain yeast proteins
yeast_mappings = stringDB_COG_mappings.loc[stringDB_COG_mappings["#protein"].str.startswith("4932", na=False),:]
yeast_mappings
```

```python
# get the yeast orthologous groups
mapping_ortho_groups = yeast_mappings["orthologous_group"].unique()


# Get the present orthologous group associations
yeast_ortho_associations = stringDB_COG_links.loc[
    stringDB_COG_links["group1"].isin(mapping_ortho_groups) | 
    stringDB_COG_links["group2"].isin(mapping_ortho_groups)
]

yeast_ortho_associations
```
