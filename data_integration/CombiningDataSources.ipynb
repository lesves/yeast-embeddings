```python
import numpy as np
import pandas as pd
```

# Previous version (only protein emb, DNA LM data and sequence features used)
- seq_feats
- lm_3utr_emb
- lm_5utr_emb
- prot_emb

```python
# DNA LM embeddings
dnalm_df = pd.read_parquet('../Data/dnalm_df.parquet')
dnalm_df.head()
```

```python
# Sequence features
seq_feats = dnalm_df[['gene_id'] + list(dnalm_df.iloc[:, 6:-1].columns)].dropna()
seq_feats.head()
```

```python
# Export seq_feats
seq_feats.to_csv("../Data/seq_feats.csv", index=True)
```

```python
# Upstream sequence embedding
lm_5utr_emb = pd.DataFrame(np.stack(dnalm_df.dropna().embedding_upstream.values), index=dnalm_df.dropna().gene_id)
lm_5utr_emb.head()
```

```python
# Downstream sequence embedding
lm_3utr_emb = pd.DataFrame(np.stack(dnalm_df.dropna().embedding_downstream.values), index=dnalm_df.dropna().gene_id)
lm_3utr_emb.head()
```

```python
# Protein embeddings from ESM2
prot_emb = pd.read_parquet('../Data/protein_emb.parquet')
prot_emb.head()
```

```python
def zscore_matrix(mat):
    m = np.mean(np.array(mat, dtype=np.float32).flatten())
    sd = np.std(np.array(mat, dtype=np.float32).flatten())

    mat_z = (mat - m)/sd
    return mat_z
```

```python
%%time

seq_feats = seq_feats.set_index('gene_id')

prot_emb_z = zscore_matrix(prot_emb)
prot_emb_z.columns = prot_emb_z.columns.astype(str)
prot_emb_z.columns = 'ESM2_' + prot_emb_z.columns
prot_emb_z = prot_emb_z.reset_index()

lm_3utr_emb_z = zscore_matrix(lm_3utr_emb)
lm_3utr_emb_z.columns = lm_3utr_emb_z.columns.astype(str)
lm_3utr_emb_z.columns = 'LM_3UTR_' + lm_3utr_emb_z.columns
lm_3utr_emb_z = lm_3utr_emb_z.reset_index()

lm_5utr_emb_z = zscore_matrix(lm_5utr_emb)
lm_5utr_emb_z.columns = lm_5utr_emb_z.columns.astype(str)
lm_5utr_emb_z.columns = 'LM_5UTR_' + lm_5utr_emb_z.columns
lm_5utr_emb_z = lm_5utr_emb_z.reset_index()
```

```python
# Merge all the embeddings to get a per gene S.cerevisiae embedding
yeast_emb_only_embeddings = seq_feats.merge(prot_emb_z, on='gene_id').merge(lm_3utr_emb_z, on='gene_id').merge(lm_5utr_emb_z, on='gene_id')
yeast_emb_only_embeddings
```

```python
# Export Dataset with only embeddings
yeast_emb_only_embeddings.to_csv("../Data/yeast_emb_only_embeddings.csv", index=True)
```

# Add Yeastnet Data

```python
yeastnet_embeddings_npz = np.load("../Data/YeastNetEmbeddings.npz")
yeastnet_embeddings = pd.DataFrame(np.stack(yeastnet_embeddings_npz["data"]), index=yeastnet_embeddings_npz["IDs"]).add_prefix("yeastnet_")
yeastnet_embeddings

yeastnet_emb_z = zscore_matrix(yeastnet_embeddings)
yeastnet_emb_z.columns = yeastnet_emb_z.columns.astype(str)
yeastnet_emb_z = yeastnet_emb_z.reset_index()
yeastnet_emb_z.rename(columns={'index':'gene_id'}, inplace=True)

yeastnet_emb_z
```

```python
yeast_emb_embeddings_yeastnet = yeast_emb_only_embeddings.merge(yeastnet_emb_z, on="gene_id", how="left")
# Fill NaN values after a left merge and ensure no genes are lost
yeast_emb_embeddings_yeastnet.fillna(yeast_emb_embeddings_yeastnet.iloc[:, 1:].mean(), inplace=True)
```

```python
# Export Dataset with embeddings + Yeastnet 
yeast_emb_embeddings_yeastnet.to_csv("../Data/yeast_emb_embeddings_yeastnet.csv", index=True)
```

# Add Data Turco Gene Embeddings

```python
data_Turco_gene_expressions = pd.read_csv('../Data/data_Turco_gen_expressions.csv', sep = ',', index_col = 0)
data_Turco_gene_expressions
```

```python
# Normalize Turco gene expressions
data_Turco_gene_expressions_z = zscore_matrix(data_Turco_gene_expressions)
data_Turco_gene_expressions_z.columns = data_Turco_gene_expressions_z.columns.astype(str)
data_Turco_gene_expressions_z.columns = 'data_Turco_expression_' + data_Turco_gene_expressions_z.columns
data_Turco_gene_expressions_z = data_Turco_gene_expressions_z.reset_index()

data_Turco_gene_expressions_z
```

```python
# Add data Turco gene expressions to the predictive model
yeast_emb_embeddings_genex = yeast_emb_only_embeddings.merge(data_Turco_gene_expressions_z, on="gene_id", how="left")
yeast_emb_embeddings_yeastnet_genex = yeast_emb_embeddings_yeastnet.merge(data_Turco_gene_expressions_z, on="gene_id", how="left")

# Fill NaN values after a left merge and ensure no genes are lost
yeast_emb_embeddings_genex.fillna(yeast_emb_embeddings_genex.iloc[:, 1:].mean(), inplace=True)
yeast_emb_embeddings_yeastnet_genex.fillna(yeast_emb_embeddings_yeastnet_genex.iloc[:, 1:].mean(), inplace=True)
```

# Add orthologous groups from STRING DB
- adding only one-hot encodings reduced the r2 by a lot !!!

```python
orthologous_groups = pd.read_csv('../Data/orthologous_groups.csv', sep = ',', index_col = 1)
orthologous_groups = orthologous_groups.drop(orthologous_groups.columns[0], axis=1)
orthologous_groups
```

```python
# Normalize Orthogroups
orthologous_groups = orthologous_groups.astype(float)
group_counts = orthologous_groups.sum(axis=1)
orthologous_groups = orthologous_groups.div(group_counts, axis=0)
```

```python
# Create dataset combinations
yeast_emb_embeddings_orthogroup = yeast_emb_only_embeddings.merge(orthologous_groups, on="gene_id", how="left")
yeast_emb_embeddings_genex_orthogroup = yeast_emb_embeddings_genex.merge(orthologous_groups, on="gene_id", how="left")
yeast_emb_embeddings_yeastnet_orthogroup = yeast_emb_embeddings_yeastnet.merge(orthologous_groups, on="gene_id", how="left")
yeast_emb_full = yeast_emb_embeddings_yeastnet_genex.merge(orthologous_groups, on="gene_id", how="left")

# Fill NaN values after a left merge and ensure no genes are lost
yeast_emb_embeddings_orthogroup.fillna(yeast_emb_embeddings_orthogroup.iloc[:, 1:].mean(), inplace=True)
yeast_emb_embeddings_genex_orthogroup.fillna(yeast_emb_embeddings_genex_orthogroup.iloc[:, 1:].mean(), inplace=True)
yeast_emb_embeddings_yeastnet_orthogroup.fillna(yeast_emb_embeddings_genex_orthogroup.iloc[:, 1:].mean(), inplace=True)
yeast_emb_full.fillna(yeast_emb_full.iloc[:, 1:].mean(), inplace=True)
```

# Extraction

```python
# Export Dataset with Embeddings + Orthologous groups
yeast_emb_embeddings_orthogroup.to_csv("../Data/yeast_emb_embeddings_orthogroup.csv", index=True)

# Export Dataset with Embeddings + Gene expressions + Orthologous groups
yeast_emb_embeddings_genex_orthogroup.to_csv("../Data/yeast_emb_embeddings_genex_orthogroup.csv", index=True)

# Export Dataset with Embeddings + Yeastnet Data + Orthologous groups
yeast_emb_embeddings_yeastnet_orthogroup.to_csv("../Data/yeast_emb_embeddings_yeastnet_orthogroup.csv", index=True)

# Extract Combined Data
yeast_emb_full.to_csv("../Data/yeast_emb_full.csv", index=True)
```
