```python
import pandas as pd
```

OrthoDB Data

Note: Data Loading and merging takes aprox. 2-3 min each since files are 3-5 gb

```python
columns_genesTab = [
    "OrthoDB_gene_id",
    "OrthoDB_organism_id",
    "Protein_sequence_id",
    "Synonyms",
    "Uniprot_id",
    "Ensembl_ids",
    "NCBI_gid_or_gene_name",
    "Description"
]

genesTab = pd.read_csv("data/odb10v1_genes.tab", sep='\t', names=columns_genesTab, low_memory=False)
genesTab
```

```python
genesTab["Description"].dropna().head(100)
```

```python
columns_OG2 = ["OG_unique_id", "OrthoDB_gene_id"]
OG2genesTab = pd.read_csv("data/odb10v1_OG2genes.tab", sep='\t', names=columns_OG2, low_memory=False)
OG2genesTab
```

```python
# Group genes by orthogroups
orthoGroups = pd.merge(OG2genesTab, genesTab, on="OrthoDB_gene_id", how="inner")
orthoGroups
```
